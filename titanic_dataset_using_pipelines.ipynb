```python

import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('/content/train (1).csv')
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
# Step 1 -> train/test/split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),
                                                 df['Survived'],
                                                 test_size=0.2,
                                                random_state=42)
```

```python
X_train.head()
```

```python
y_train.sample(5)
```

```python
# imputation transformer
trf1 = ColumnTransformer([
    ('impute_age',SimpleImputer(),[2]),
    ('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])
],remainder='passthrough')
```

```python
# one hot encoding
trf2 = ColumnTransformer([
    ('ohe_sex_embarked',OneHotEncoder(handle_unknown='ignore'),[1,6])
],remainder='passthrough')
```

```python
# Scaling
trf3 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])
```

```python
# Feature selection
trf4 = SelectKBest(score_func=chi2,k=8)
```

```python
# train the model
trf5 = DecisionTreeClassifier()
```

Create Pipelines

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),

    ('trf4',trf4),
    ('trf5',trf5)
])
```

Pipelinevs make_pipeline

(Same applies at the column vs make_column_transformer

```python
#Alternate Syntax
pipe = make_pipeline  (trf1,trf2,trf3,trf4,trf5)
```

```python
# train
pipe.fit(X_train,y_train)
```

 Pipeline ka Basic Kaam Dataset me

Data Cleaning (Preprocessing)

Missing values ko handle karta hai (mean, median, most frequent).

Example: Agar "Age" column me kuch values missing hai  pipeline automatically mean dal dega.

Encoding

Categorical data (jaise "Male", "Female", "City = Delhi") ko numbers me convert karta hai.

Example: Male  0, Female  1 ya OneHotEncoding.

Scaling / Normalization

Numerical values ko ek proper range me le aata hai (01 ya -11).

Example: Salary (20k  2L) ko normalize karke model ke liye easy banata hai.

Feature Selection / Dimensionality Reduction

Un features ko choose karta hai jo model ke liye zyada useful hain.

Example: SelectKBest sirf top k features rakhta hai.

Model Training

Last step me koi algorithm (Decision Tree, Logistic Regression, etc.) lagata hai aur dataset pe training karta hai.

Prediction with Same Steps

Jab naya data aata hai, pipeline automatically wahi saare steps (cleaning  encoding  scaling  selection) apply karke prediction de deta hai.
