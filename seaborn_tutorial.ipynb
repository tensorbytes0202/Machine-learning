```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

1.Relational
* to see the statistical relation between 2 or more variables
* bivariate analysis

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.sample(10)
```

```python
#scatter plot
sns.scatterplot(data=tips,x='total_bill',y='tip')
#agar matplotlib me karte to uske pura batana padta ki x-axis ko kya label dena h y-axis ko kya or title wagera sab
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter')
```

scatterplot() → Sirf ek scatter plot banane ke liye hota hai.

relplot() → Ek higher-level function hai jo scatter ya line dono banata hai + facet (subplots) banane ke liye powerful hai.

👉 Matlab simple case me scatterplot use karo, aur agar multiple plots ek saath banana hai (subplots ya categories ke hisaab se), tab relplot use karo.

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

#lineplot
Lineplot ka use tab hota hai jab tumhe kisi cheez ka “trend ya change” dikhana ho time ke sath ya kisi continuous variable ke sath.

px.data.gapminder() ek sample dataset hai jo Plotly Express library ke andar built-in aata hai.

country → Country ka naam

continent → Country kis continent me hai

year → Data kis saal ka hai

lifeExp → Life expectancy (average age people live)

pop → Population

gdpPercap → GDP per capita (per person income)

```python
gap = px.data.gapminder()
gap
```

```python
temp_df=gap[gap['country'] == 'India']
temp_df
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
#usind relplot
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
#hue->style
temp_df = gap[gap['country'].isin(['India','Pakistan','China'])]
temp_df
```

```python
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue='country')
```

the difference between the line plot and the relplot ki lineplot apni axis tak rehta h bhar nhi jata jabki relplot out the range nikal jata h

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp',hue='country')
```

```python
#hue->style
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
temp_df.sample(15)
```

```python
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue='country')
```

```python
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue='country',style='continent')
```

#Facetplot

Distribution plot:
Output me kya dikhega?

Histogram (bars) → dikhata hai kitni values ek particular range me aayi hain.

KDE Curve (optional, kde=True) → ek smooth curve jo distribution ka shape dikhata hai (bell shape, skewed, etc.).

Bins → decide karte hain kitne intervals me data ko todna hai.

✅ Use of Distribution Plot:

Data ka spread samajhna (values kaise banti hain).

Data normal distribution follow karta hai ya nahi dekhna.

Outliers identify karne ke liye.


1.histplot
2.kdeplot
3,rugplot


figure level=> distplot
axes level -> hist,kde,rugplot

```python
#plot a univariate hist plot
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist',bins=20)
```

A rugplot is a simple visualization tool in Seaborn (Python) that shows the distribution of a dataset by drawing small vertical (or horizontal) lines (like ticks) along an axis for each data point.

```python
#kdeplot
# KDE Plot (Kernel Density Estimate Plot)
# y-axis density probablity hota h

# KDE Plot ek smooth curve hota hai jo data ki probability distribution dikhata hai.
# Iska kaam hai histogram ke jagah ek continuous smooth line dikhana jisse distribution ka pattern easily samajh aaye.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex')
```

```python
sns.histplot(data=tips,x='total_bill',y='tip')
#bivariate using two features
```

```python
sns.kdeplot(data=tips,x='total_bill',y='tip')
```

```python
#Heatmap := it is the axes level function
#Clustermap
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(10,10))
sns.heatmap(data=temp_df)
```

```python
sns.clustermap(data=temp_df)
plt.figure(figsize=(15,20))
```

```python
gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
```

```python
temp_df=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
sns.heatmap(data=temp_df)
```

Annotation


```python
sns.clustermap(data=temp_df,annot=True)
plt.figure(figsize=(10,15))
```

Line width function

```python
sns.clustermap(data=temp_df,annot=True,linewidth=0.5)
```

```python
sns.clustermap(data=temp_df,annot=True,linewidth=0.5,cmap='Blues')
```

```python
sns.clustermap(data=temp_df,annot=True,linewidth=0.5,cmap='viridis')
```
