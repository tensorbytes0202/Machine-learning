```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Mixed data.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
df.isnull().sum()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passengers travelling with')
```

```python
# extract numerical part
df['number_numerical'] = pd.to_numeric(df["number"],errors='coerce',downcast='integer')
```

to_numeric() pandas ka function hai jo kisi bhi column ko numeric (integer ya float) me convert karta hai.

errors='coerce'

Agar conversion ke dauran koi value numeric me convert nahi ho sakti (jaise text), to coerce ka matlab hai us value ko NaN (Not a Number) bana do.

Isse errors avoid ho jati hain.

downcast='integer'

Ye batata hai ki result ko possible ho to smallest integer type me convert karo.

Example: agar values chhoti hain to int64 ki jagah int32 ya int16 me store karo memory efficient banane ke liye.

```python
# extract categorical part
df['number_categorical'] = np.where(df['number_numerical'].isnull(),df['number'],np.nan)

df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)') # captures numerical part
df['cabin_cat'] = df['Cabin'].str[0] # captures the first letter

df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
